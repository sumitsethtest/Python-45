# 함수의 return

앞서 설명한 것과 마찬가지로 함수는 반환되는 값이 있으며, 이는 어떠한 종류의 객체여도 상관없습니다. 
 
단, 오직 한 개의 객체만 반환됩니다.

**함수가 return 되거나 종료되면, 함수를 호출한 곳으로 돌아갑니다.**




## 함수를 정의하고 값을 반환해봅시다.

> 함수는 모든 객체를 리턴할 수 있습니다. 
> 
> 리스트 두개를 받아 각각 더한 결과를 비교하여 값이 큰 리스트를 반환합니다.

```python
my_list_max([10, 3], [5, 9])
```
---
```
예상 출력)
[5, 9]
```


````python
# 여기에 my_list_max 함수를 만들어주세요.
def my_list_max(a, b):
    return a if sum(a) > sum(b) else b 
r = my_list_max([10, 3], [5, 9])
print(r)

# 실수: 변수에 sum(a), sum(b)를 담아서 결과로 비교했기 때문에 copy가 된 상태여서 리스트로 출력되지 않았음. 
# **함수가 return 되거나 종료되면, 함수를 호출한 곳으로 돌아갑니다.**
````

````python
def my_list_max(a, b):
    if sum(a) > sum(b):
        return a
    else:
        return b 
r = my_list_max([10, 3], [5, 9])
print(r)
````

## 가변 인자 리스트

앞서 설명한 `print()`처럼 개수가 정해지지 않은 임의의 인자를 받기 위해서는 가변인자를 활용합니다. 

가변인자는 `tuple` 형태로 처리가 되며, `*`로 표현합니다. 

**활용법**

```python
def func(*args):
```

````python
objects = 'tt'
print(objects)
print(*objects, sep='//') # 't' 't' 로 인식

shingihea = '123'
print(shingihea)
print(*shingihea, sep='//')
````

````python
# args는 tuple!
def my_func(*args):
    print(args)
    print(type(args))
    
my_func(123, 456, 765, 342)
````

````python
def my_func(*args, second_arg): # second arg로 positional 하게 approach할 수 없으므로,
    print(args)
    print(type(args))
    print(second_arg)
    
my_func(123, 456, 765, 342)
````

````python
def my_func(*args, second_arg='second'): # keyword로 approach해야한다
    print(args)
    print(type(args))
    print(second_arg)
    
my_func(123, 456, 765, 342, second_arg='식사 맛있게 하세요.')
````

###  URL 편하게 만들기

> url 패턴을 만들어 문자열을 반환하는  `my_url` 함수를 만들어봅시다.
>
> 영진위에서 제공하는 일별 박스오피스 API 서비스는 다음과 같은 방식으로 요청을 받습니다.

```
기본 요청 URL : http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?
```

* key : 발급받은 키값(abc)
* targetDt : yyyymmdd
* itemPerPage : 1 ~ 10 **기본 10**


```
예시)
my_url(key='abc', targetDt='yyyymmdd')

api = {
    'key': 'abc',
    'targetDt': 'yyyymmdd'
}
my_url(**api)

예시 출력)
'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?itemPerPage=10&key=abc&targetDt=yyyymmdd&'
```

````python
# 여기에 코드를 작성해주세요.
def my_url(key, targetDt, itemPerPage='10'):
    base_url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?'
    
    itemPerPage = int(itemPerPage)
    if 1 <= itemPerPage <= 10:
        return f'{base_url}itemPerPage={itemPerPage}&key={key}&targetDt={targetDt}&'
    else: 
        return f'page는 1~10중에서 입력해주세요.'

# 1.
api = {
    'key': 'abc',
    'targetDt': 'yyyymmdd',
}

print(my_url(**api))

# 2.
print(my_url(key='abc', targetDt='yyyymmdd'))
````

````python
def my_url_second(itemPerPage='10', **kwargs):
    base_url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?'
    print(kwargs, itemPerPage)
    for key, val in kwargs.items():
        print(key, val)
    return f'{base_url}itemPerPage={itemPerPage}&key={key}&targetDt={val}&'

print(my_url_second(key='abc', targetDt='yyyymmdd', itemPerPage=5))
````

### URL 검증하기

> 이제 우리는 만들어진 요청 보내기전에 URL을 검증해야합니다. 
>
> 앞선 설명을 참고하여 검증 로직을 구현하고 문자열을 반환하세요.

```
> 아래의 두가지 상황만 만들도록 하겠습니다. <

key, targetDt가 없으면, '필수 요청변수가 누락되었습니다.'

itemPerPage의 범위가 1~10을 넘어가면, '1~10까지의 값을 넣어주세요.'
```

````python
# 여기에 코드를 작성해주세요
def my_url_second(itemPerPage='10', **kwargs):
    base_url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?'
    print(kwargs, itemPerPage)
    
    #check_keynval = []
    
    #if 'key' not in check_keynval and 'targetDt' not in check_keynval:
    #    return '필수 요청변수가 누락되었습니다.'
    
    for key, val in kwargs.items():
        print(key, val)
        #check_keynval.append(key)
        #print(check_keynval)
    
    #check complete
    if 1 <= int(itemPerPage) <= 10:
        return f'{base_url}itemPerPage={itemPerPage}&key={kwargs["key"]}&targetDt={kwargs["targetDt"]}&'
    else:
        return '1~10까지의 값을 넣어주세요.'

print(my_url_second(key='abc', targetDt='yyyymmdd', itemPerPage=5))
````

````python
# 여기에 코드를 작성해주세요
def my_url_second(itemPerPage='10', **kwargs):
    base_url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?'

    list_key = []
    list_val = []
    
    for key, val in kwargs.items():
        list_key.append(key)
        list_val.append(val)

    if 'key' not in list_key or 'targetDt' not in list_key:
        return '필수 요청변수가 누락되었습니다.'
    
    if 1 <= int(itemPerPage) <= 10:
        return f'{base_url}itemPerPage={itemPerPage}&{list_key[0]}={list_val[0]}&{list_key[1]}={list_val[1]}&'
    else:
        return '1~10까지의 값을 넣어주세요.'

print(my_url_second(key='abc', targetDt='yyyymmdd', itemPerPage=5))
````

````python
def my_url_second(itemPerPage='10', **kwargs):
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?'
    url = url + f'itemPerPage={itemPerPage}&'
    # print(url)
    # print(kwargs) # {'key': 'abc', 'targetDt': 'yyyymmdd'} 딕셔너리 형태로 들어온다.
    
    for key, value in kwargs.items():
        url = url + f'{key}={value}&'
    
    return url
    
print(my_url_second(key='abc', targetDt='yyyymmdd', itemPerPage=5))
````

## 이름공간 및 스코프(Scope)

파이썬에서 사용되는 이름들은 이름공간(namespce)에 저장되어 있습니다.
그리고, LEGB Rule을 가지고 있습니다. 

변수에서 값을 찾을 때 아래와 같은 순서대로 이름을 찾아나갑니다.
* `L`ocal scope: 정의된 함수
* `E`nclosed scope: 상위 함수 
* `G`lobal scope: 함수 밖의 변수 혹은 import된 모듈
* `B`uilt-in scope: 파이썬안에 내장되어 있는 함수 또는 속성

```
반복문 코드:

    재귀 함수를 호출하며, n은 1씩 감소한다.
```
